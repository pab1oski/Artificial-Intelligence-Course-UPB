### Slicing Strings

The string slice function creates substrings by following the rule:

```python
string_object[start_index:end_index:step]
```

Besides,

```python
string_object[:index] + string_object[index:] == string_object
```

````python
s = 'Practice makes perfect!'
first_word = s[:8]
print(first_word)
first_word_reversed = first_word[::-2]
print(first_word_reversed)
````

````python
def do_stuff(my_str, start1, end1, start2, end2):
    """This function extracts
    the required slices from 
    a given string passed in 
    as a parameter and returns
    a new string.
    """
    new_str1=my_str[start1:end1]
    new_str2=my_str[start2:end2]
    return '{}{}'.format(new_str1,new_str2)

if __name__ == '__main__':
    str_arg = 'The AI journey is the perfect opportunity to expand one`s horizons.'
    # Call function.
    string=do_stuff(str_arg,4,7,len(str_arg)-9,len(str_arg))
    # The expected new string is 'AI horizons.'
    assert string=='AI horizons.'
````

**c.** Calculate the number of upper case letters, lower case letters and punctuation marks used in a given sentence.
For the string used before, the output should be:

`UPPER CASE 3, lower case 52 and punctuation marks 2.`

_Note_: A set of built-in methods that you can use on strings are `str.upper()`, `str.isupper()` ,`str.lower()`, `str.islower()`, `str.title()`.

````python
def compute_characters(sentence):
    ### Your implementation here:
    d={
        "lower_ch":0,
        "upper_ch":0,
        "puncts":0
    }
    for character in sentence:
        if character.islower():
            d["lower_ch"]+=1
        elif character.isupper():
            d["upper_ch"]+=1
        elif character.isspace()==False:
            d["puncts"]+=1     
                  
    
    ###
    
    print("UPPER CASE {}, lower case {} and punctuation marks {}." .format(d["upper_ch"], d["lower_ch"], d["puncts"]))

my_str = '.'
print(my_str)

compute_characters(str_arg)
````

### Multiple Arguments `(*args)` and keyword arguments `(**kwargs)`

_Note:_ Used to make the function flexible.

````python
def compute_sum(a, b):
    return a + b
print(compute_sum(4, 5))
````

````python
def compute_sum(*args):
    print('Type of iterable object args {}.' .format(type(args)))
    count = 0
    for a in args:
        count += a
    return count
````

````python
result = compute_sum(0.5, 2.25, 7)
print('Result: ', result)
````

**Obs:** The `args` argument is a name, the key-element is the unpacking operator `*`.

### Task 3

The following formula computes a _Z score_ and measures how far a single raw data value is from the population mean.

\begin{equation*}
z = \frac{X - \mu }{\sigma }
\end{equation*}

where:
* **_X_** is a single raw data value
* `mu` is the population mean
* `sigma` is the population standard deviation

To find the standard deviation, the equation below comes in hand:

\begin{equation*}
\sigma = \sqrt{\frac{\sum \left | X - \mu \right |^{2}}{N}}
\end{equation*}

where **_N_** is the number of data points in the population.

**a.** Using `sum()` and `list comprehension`, compute the mean and the standard deviation for the population defined below:

````python
data =  [4.5, 5, 5.5, 6, 6.25, 7, 15.25, 18, 18.45, 21, 21.45, 23]
print(data)
````

````python
# Your implementation here:
mean =sum(data)/len(data)
std=(sum([(num-mean)**2 for num in data])/len(data))**0.5
print(mean)
print(std)




````

**b.** Define the `z_score()` function and implement the mathematical expression. The obtained values should be stored in a _z score_ values list and rounded to 3 decimals.

````python
# Your implementatio here:
def z_score(data,mean,std):
    Z=[(num-mean)/std for num in data]
    return Z


````

**c.** Add the corresponding elongation of each raw data value into a dictionary.

````python
# Your implementatio here:
data_trans=z_score(data,mean,std)
data_dict={}
for i in range(len(data)):
    data_dict[data[i]]=data_trans[i]

print(type(data_dict))
print(data_dict)
````
