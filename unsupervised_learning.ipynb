### Short Demo

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Values extracted from the official scikit-learn documentation (check the first link).
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])
```

```python
X
```

```python
# Visualise data
plt.scatter(X[:,0], X[:,1])
```

```python
# Agglomerative Hierarchical Clustering model with no parameters
aggl_model_demo = AgglomerativeClustering().fit(X)
```

```python
aggl_model_demo
```

```python
# Cluster labels for each point
aggl_model_demo.labels_
```

```python
# Visualise data

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[aggl_model_demo.labels_,0], X[aggl_model_demo.labels_,1], c='r')
```

### Task 2

Try to modify the number of clusters for `data2.txt` and write down your observations for the _Agglomerative Hierarchical Clustering_ model.

```python
# Load and see the data
data=np.loadtxt('data2.txt', usecols=(0, 1), skiprows=1)

plt.scatter(data[:,0],data[:,1])


## We can see that the number of cluster are no well-defined
```

```python
# As we cannot see as ina firt view the optimal 
# number of clusters , we use the elbow rule

from yellowbrick.cluster import KElbowVisualizer

model=AgglomerativeClustering()

visualizer=KElbowVisualizer(model,k=(1,15))
visualizer.fit(X)
visualizer.show()
n_cluster=visualizer.elbow_value_
print("The optimal number of cluster is:",n_cluster)
```

```python
# Fit the model with the optimal number of clusters
model=AgglomerativeClustering(n_clusters=n_cluster)
model.fit(data)
labels=model.labels_
plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis')
```
